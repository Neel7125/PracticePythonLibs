```python
# Load dataset
df = pd.read_csv('us_presidents.csv')

# Display frst 10 rows
print("First 10 rows:\n", df.head(10))

print('-'*50)

# Display last 10 rows
print("Last 10 rows:\n", df.tail(10))
```

### Data Cleaning

```python
# Drop columns

if "prior" in df.columns:
    df_cleaned = df.drop(columns=['prior'])
    print("\n-------Dropped 'prior' Column-------\n", df_cleaned.head(10))
else:
    print("\n'prior' column not found in the Dataset.")
```

```python
# Fill missing values

if "end" in df.columns:
    df["end"].fillna("Present", inplace=True)
```

```python
# Convert date columns to datetime

for col in ['start', 'end']:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

print("Converted DateTime\n")
print(df[['president', 'start', 'end']].head(50))
```

```python
# Create new column: presidency duration (in days)

if 'start' in df.columns and 'end' in df.columns:
    df['duration_days'] = (df['end'] - df['start']).dt.days

print("Duration :\n")
print(df[['president', 'start', 'end', 'duration_days']])
```

```python
# Save cleaned dataset

df.to_csv('us_presidents_cleaned.csv', index=False)
print("✅ Cleaned dataset saved to 'us_presidents_cleaned.csv'")
```

### Grouping & Aggregation

```python
# Group by party → count presidents

print("Number of Presidents by Party:\n")
print(df.groupby("party")['president'].count())
```

```python
# Group by party → average presidency duration

if 'duration_days' in df.columns:
    print("Average Presidency Duration by Party:\n")
    print(df.groupby("party")['duration_days'].mean())
```

```python
# Value counts (how many from each party)

print("Value Counts by Party:\n")
print(df['party'].value_counts())
```

```python
# Sorting by presidency duration (longest → shortest)

if 'duration_days' in df.columns:
    print("Sorted Presidendicy Duration:\n")
    print(df.sort_values(by='duration_days', ascending=False)[['president', 'duration_days']])
```

```python
# Sorting by start year

if 'start' in df.columns:
    print("Sorted by Start Year: \n")
    print(df.sort_values(by='start', ascending=False)[['president', 'start']])
```

### String Operations

```python
# Check if "Prior" column contains "Vice President"

df['Was_VP'] = df['prior'].str.contains('Vice President', case=False, na=False)
print("President who were Vice Presidents:\n")
print(df[['president', 'Was_VP']])
```

```python
# Extract year from Start / End

if 'start' in df.columns:
    df['start_year'] = df['start'].dt.year
if 'end' in df.columns:
    df['end_year'] = df['end'].dt.year

print("Extracted Years:\n")
print(df[['president', 'start_year', 'end_year']])
```

```python
# Clean extra spaces / characters in Party column

if 'party' in df.columns:
    df["party_clean"] = df['party'].str.strip().str.replace(r'[^a-zA-Z ]', "", regex=True)
    print("Cleaned Party Names:\n")
    print(df[['president','party_clean']])
```

### Merging & Exporting

```python
# create another small dataset (US states of birth for some presidents)

data_states ={
    "president": ["George Washington", "Jhon F. Kennedy", "George W. Bush", "Barack Obama", "Donald Trump"],
    "birth_state": ["Virginia", "Massachusetts", "Connecticut", "Hawaii", "New York"]
}
df_states = pd.DataFrame(data_states)

print("Dataset of States of Birth:\n")
print(df_states)
```

```python
# Merge with presidents dataset on Name

df_merged = pd.merge(df, df_states, on='president', how='left')
print("Merged Dataset:\n")
print(df_merged[['president', 'birth_state']])
```

```python
# Export final merged dataset

df_merged.to_csv("us_presidents_merged.csv", index=False)
print("✅ Merged dataset saved to 'us_presidents_merged.csv'")
```

### Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bar chart: presidents per party

plt.figure(figsize=(10,6))
sns.countplot(data=df, x='party', order=df['party'].value_counts().index)
plt.title("Number of Presidents by Party")
plt.xticks(rotation=45)
plt.show()
```

```python
# Timeline: presidency durations

plt.figure(figsize=(10,5))
for i, row in df.iterrows():
    if pd.notnull(row['start']) and pd.notnull(row['end']):
        plt.plot([row['start'],row['end']], [row["president"], row["president"]], marker='o')
plt.title("Presidency Durations")
plt.xlabel("Year")
plt.ylabel("President")
plt.show()
```

```python
# Pie chart: distribution of presidents by party

plt.figure(figsize=(10,5))
df['party'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, cmap='tab20')
plt.ylabel('')
plt.title('Distribution of Presidents by Party')
plt.show()
```

```python
# Countplot: presidents by century

df["century"] = df['start'].dt.year // 100 + 1
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="century", palette="coolwarm")
plt.title("Number of Presidents by Century")
plt.show()
```
